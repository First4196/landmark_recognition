```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("train/train.csv")
train_label_to_category = pd.read_csv("train/train_label_to_category.csv")
```

```python
train = train.groupby('landmark_id').filter(lambda x : len(x) >= 500)
train = train.merge(train_label_to_category, on='landmark_id', how='inner')
train['category'] = train['category'].str.replace('http://commons.wikimedia.org/wiki/Category:', '').str.lower()
```

```python
keywords = 'mount|lake|fall|cave|canyon|park|garden|cemetery|castle|palace|monastery|museum|fort|port|tower|library|monument|shrine|church|bridge'
train_selected = train[train['category'].str.contains(keywords, regex=True) & ~train['category'].str.contains('%')]
train_selected = train_selected.reset_index(drop=True)
train_selected['category'].value_counts()
```

```python
train_selected.head()
```

```python
train_selected_label_to_category = train_selected[['landmark_id', 'category']].drop_duplicates()
train_selected_label_to_category = train_selected_label_to_category.reset_index(drop=True)
train_selected_label_to_category['label'] = train_selected_label_to_category.index
train_selected_label_to_category.head()
```

```python
train_selected = train_selected.merge(train_selected_label_to_category[['landmark_id', 'label']], on='landmark_id', how='left')
train_selected = train_selected.reset_index()
train_selected.head()
```

```python
train_selected = train_selected.drop(columns=['id', 'landmark_id', 'category'])
train_selected.head()
```

```python
train_selected.to_csv('data/data.csv', index=False)
train_selected_label_to_category.to_csv('data/label_to_category.csv', index=False)
```
